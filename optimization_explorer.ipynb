```python
def save_figures(data_folder, figure_name):
    f = sns.plt.gcf()
    plot_folder = os.path.abspath(join(data_folder, "plots"))
    f.savefig("{}/{}".format(plot_folder, figure_name))

def rand_jitter(arr,amt): # 0.002<amt<0.01
    stdev = amt*(max(arr)-min(arr))
    return arr + np.random.randn(len(arr)) * stdev

def subplots(df1, df2, mainAnno, anno, color, dim1, dim2, ylim, defaults, Title, jitter):
    f, axarr = plt.subplots(dim1, dim2, figsize=(4*dim2, 3.5*dim1))
    k=0
    for i, j in [(i,j) for i in range(dim1) for j in range(dim2)]:
        axarr[i, j].scatter(rand_jitter(df2[anno[k]],jitter), df1[mainAnno], 
                            marker='o', alpha=0.4, c=df2[color], label=df2[anno[0]])
        # axarr[i, j].scatter(df2[anno[k]], df2[mainAnno])
        axarr[i, j].set_xlabel(anno[k])
        axarr[i, j].set_ylabel(mainAnno)
        axarr[i, j].axvline(x=defaults[k],linewidth=1, color='black',linestyle='dashed')
        axarr[i, j].axhline(y=df1.iloc[0][mainAnno],linewidth=1, color='black',linestyle='dashed')
        # axarr[i, j].axhline(y=df2.iloc[0][mainAnno],linewidth=1, color='g')
        axarr[i, j].set_ylim(ylim)
        k = k+1
    f.suptitle(Title, fontsize=20)
   # plt.legend(fontsize=10,loc='best')#bbox_to_anchor=(0.62, 0.9), loc=2, ncol=1, borderaxespad=0.)
    plt.setp([a.get_xticklabels() for a in axarr[0, :]])
    plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False)
    plt.tight_layout()
    plt.subplots_adjust(top=0.94)
    
def save_figures(output_folder, figure_name):
    figure = sns.plt.gcf()
    plot_folder = os.path.abspath(join(data_folder, output_folder, "plots"))
    if not os.path.isdir(plot_folder): os.mkdir(plot_folder)
    figure.savefig("{}/{}".format(plot_folder, figure_name))

plot_folder = os.path.abspath(join(data_folder, "plots"))
if not os.path.isdir(plot_folder):
             os.mkdir(plot_folder)
```

# Fitting

```python
## Optimization fitting using Ian's scripts

param_accuracy_sorted = cell_df_paramgrouped_mean.sort('align_percent_identical',ascending=False)
columns = ['align_percent_identical','p_anti','p_dwell','max_oc_length']
param_accuracy_df = param_accuracy_sorted[columns]

X_train = param_accuracy_df.values[:,1:]
y_train = param_accuracy_df['align_percent_identical'].values
# reg = linear_model.LinearRegression()
# reg = linear_model.Ridge(alpha = 0.5)
reg = linear_model.Lasso(alpha = 0.1)
#reg = linear_model.BayesianRidge()
reg.fit(X_train,y_train)

# The coefficients
print('Coefficients: \n', reg.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % np.mean((reg.predict(X_train) - y_train) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % reg.score(X_train, y_train))

from sklearn import ensemble
from sklearn import datasets
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error
params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls'}
clf = ensemble.GradientBoostingRegressor(**params)

clf.fit(X_train, y_train)
mse = mean_squared_error(y_train, clf.predict(X_train))
print("MSE: %.4f" % mse)
# compute test set deviance
test_score = np.zeros((params['n_estimators'],), dtype=np.float64)

for i, y_pred in enumerate(clf.staged_predict(X_train)):
    test_score[i] = clf.loss_(y_train, y_pred)

plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
plt.title('Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, clf.train_score_, 'b-',
         label='Training Set Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, test_score, 'r-',
         label='Test Set Deviance')
plt.legend(loc='upper right')
plt.xlabel('Boosting Iterations')
plt.ylabel('Deviance')

feature_importance = clf.feature_importances_
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.figure(figsize=(30,10))
plt.subplot(1, 2, 2)
plt.barh(pos, feature_importance[sorted_idx], align='center')
feature_names = np.array(columns[1:])
plt.yticks(pos, feature_names[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
# plt.rcParams["figure.figsize"] = (10,10)
plt.show()
```
